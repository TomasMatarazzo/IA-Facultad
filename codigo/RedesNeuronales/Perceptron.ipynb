### Perceptron

#### Como calculamos el valor de los pesos?
 
 El objetivo es buscar un valor de los pesos que permita crear un modelo que clasifique correctamente los datos,
 para ello inicializamos en este caso los pesos en cero, verificamos el error en este punto , si el error es > 0
 actualizamos el valor de los pesos, esto lo hacemos iterativamente hasta que todos los datos de test no tengan error.
 
 ##### Error = SalidaObtenida - SalidaDeseada







```python
import numpy as np
import matplotlib.pyplot as plt

#Datos que tome prestados de un youtuber ;)

# Datos de 10 personas -> [edad, ahorro]

        # Bias i1 		i2 		y
x = np.array([
        [1.00,2,1],
        [2,3,1],
        [3,4,1],
        [1,3,1],
        [2.5,2,1],
        [3,3,1],
        [4,3,1]])

clases = [-1,-1,-1,-1,1,1,1]

weights= [0.20,1.00,  -1.00] # initial weights specified in problem

plt.scatter(x[:,0],x[:,1])
```

```python
class Perceptron:
    def __init__(self,alpha, data , clases):
        '''Inicializo pesos en zero'''
        self.pesos = [0,0,0]
        self.data = data
        self.n_entradas, self.n_features = data.shape
        self.alpha = alpha
        self.clases = clases
        
    def f(self, x):
        print(x)
        return 1 if x >= 0 else -1
        
    def activacion(self , entradas):
        '''Calcula la suma ponderada
           w1*x1 + w2*x2 + ⋯ + wn*xn
        '''
        print('entradasss\n', entradas)
        y = np.dot(self.pesos,entradas)
        self.salida = self.f(y)
        print('salida \n',self.salida)
        
    def actualizacionPesos(self, entrada ,salidaEsperada):
        print(salidaEsperada)
        error = salidaEsperada - self.salida
        print('error es \n ',error)
        '''solamente actualizo los pesos , si la clasificacion es incorrecta'''
        if (error != 0):
            print(self.pesos,self.alpha,error,entrada)
            self.pesos = self.pesos + self.alpha*error*entrada
        print('nuevos', self.pesos)
            
    def calcular(self):
        for i in range(3000):
            pesosViejos = self.pesos
            for i in range(self.n_entradas):
                self.activacion(self.data[i])
                self.actualizacionPesos(self.data[i],self.clases[i])
            if (np.array_equal(pesosViejos,self.pesos):
                break
```

```python
perceptron = Perceptron(0.5,x,clases)
```

```python
perceptron.calcular()
```
